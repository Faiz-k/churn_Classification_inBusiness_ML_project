```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
path="C:\\Users\\mf879\\OneDrive\\Desktop\\44_customerchurnClassification\\Churn_Modelling.csv"
data=pd.read_csv(path)
categorical_columns = ['Surname', 'Geography', 'Gender']
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    data[col + '_new'] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
data
```

```python
inputs=data.drop(['RowNumber','Surname', 'Geography', 'Gender','Exited','CustomerId','Surname','Surname_new'],'columns')
outputs=data.drop(['RowNumber','CustomerId','Surname','CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Surname_new','Geography_new','Gender_new'],'columns')
```

```python
inputs
```

```python
outputs
```

```python
import sklearn 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
model=KNeighborsClassifier(n_neighbors=13)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
```
